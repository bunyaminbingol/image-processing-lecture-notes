* **Histogram ile resmin değer yoğunlukları ile alakalı bilgileri elde edebiliriz. Yni bir resmin histogramına bakınca resmin aydınlık noktaları, karanlık noktaları ve konstrat değerleri ile ilgili çıkarımlar yapabiliriz.**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
img = np.zeros((500,500), np.uint8) # Resmimizi oluşturuyoruz

plt.hist(img.ravel(), 256, [0,256]) # Histogramımızı çizdiriyoruz.
# 1. parametre resmimizin tüm piksel değerlerini alıp tek bir satır haline getiriyor.
# 2. parametre kaç değer olduğu
# 3. parametre değer aralığı
plt.show()

cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**(500,500) bir matrisimiz olduğu için y ekseni 250000 çıkıyor, x ekseni ise, çizdiğimiz resim tek renk(siyah) olduğu için 0 olarak kalıyor.**

```python
img = cv2.imread("veriler/smile.jpg")

plt.hist(img.ravel(), 256, [0,256])
plt.show()

cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# BGR değerlerinin histogramını çıkartalım:

img = cv2.imread("veriler/smile.jpg")
b,g,r = cv2.split(img) # Resmin BGR değerlerini okuyalım

plt.hist(b.ravel(), 256, [0,256], color = "b")
plt.hist(g.ravel(), 256, [0,256], color = "g")
plt.hist(r.ravel(), 256, [0,256], color = "r")
plt.show()

cv2.imshow("Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
