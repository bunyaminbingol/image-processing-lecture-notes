* **Bitwise işleçleri yaparken iki resim kullanacağız bu resimler şunlardır:  bu resimleri üst üste oturtacağız ve bitwise işlemleri yapacağız. Resimleri matris olarak düşünürsek 1 ve 0 lardan oluşuyor Siyah kısımlar 0, Beyaz kısımlar 1. Bitwise işlemlerde de resmin 0 ve 1'leri karşılaştırılıyor buna göre bir sonuç çıkartılıyor. Bu fonksiyonda mantıksal işlemleri kullanacağız.**
* Bit düzeyinde işlem yapan operatörlerdir; yani sayıları bir bütün olarak değil, bit bit ele alarak işleme sokarlar. Bit operatörlerinin aldığı operandların tamsayı türlerinden olması gerekir (char,short,int,long),float yada double türleri bu operatörlerle işleme sokulamaz.Bit operatörlerinin listesi,”kendi aralarındaki öncelik sırası gözönüne alınarak” aşağıda verilmiştir:


```python
import cv2
import numpy as np

img1 = cv2.imread("veriler/bitwise_1.png")
img2 = cv2.imread("veriler/bitwise_2.png")
```

```python
# Şimdi bu resimlere bitwise işlemleri uygulayacağız ve sonuçlarını tek tek gözlemleyeceğiz.
# İlk olarak "and" işlemine tabi tutalım

bit_and = cv2.bitwise_and(img2, img1) # Bu fonksiyon bit düzeyinde "and" işlemlerini gerçekleştirecek

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_and", bit_and)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# "or" işlemine tabi tutalım

bit_or = cv2.bitwise_or(img2, img1) # Bu fonksiyon bit düzeyinde "or" işlemlerini gerçekleştirecek

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_or", bit_or)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# "xor" işlemine tabi tutalım

bit_xor = cv2.bitwise_xor(img2, img1) # Bu fonksiyon bit düzeyinde "xor" işlemlerini gerçekleştirecek

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_xor", bit_xor)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# "not" işlemine tabi tutalım

bit_not1 = cv2.bitwise_not(img1) # Bu fonksiyon bit düzeyinde "not" işlemlerini gerçekleştirecek
bit_not2 = cv2.bitwise_not(img2)

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_not1", bit_not1)
cv2.imshow("bit_not2", bit_not2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
