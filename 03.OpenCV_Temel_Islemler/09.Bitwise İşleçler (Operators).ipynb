Bit operatrlerinin ald operandlarn tamsay trlerinden olmas gerekir (char,short,int,long),float yada double trleri bu operatrlerle ileme sokulamaz.Bit operatrlerinin listesi,kendi aralarndaki ncelik sras gznne alnarak aada verilmitir:


```python
import cv2
import numpy as np

img1 = cv2.imread("veriler/bitwise_1.png")
img2 = cv2.imread("veriler/bitwise_2.png")
```

```python
# imdi bu resimlere bitwise ilemleri uygulayacaz ve sonularn tek tek gzlemleyeceiz.
# lk olarak "and" ilemine tabi tutalm

bit_and = cv2.bitwise_and(img2, img1) # Bu fonksiyon bit dzeyinde "and" ilemlerini gerekletirecek

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_and", bit_and)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# "or" ilemine tabi tutalm

bit_or = cv2.bitwise_or(img2, img1) # Bu fonksiyon bit dzeyinde "or" ilemlerini gerekletirecek

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_or", bit_or)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# "xor" ilemine tabi tutalm

bit_xor = cv2.bitwise_xor(img2, img1) # Bu fonksiyon bit dzeyinde "xor" ilemlerini gerekletirecek

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_xor", bit_xor)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# "not" ilemine tabi tutalm

bit_not1 = cv2.bitwise_not(img1) # Bu fonksiyon bit dzeyinde "not" ilemlerini gerekletirecek
bit_not2 = cv2.bitwise_not(img2)

cv2.imshow("img1", img1)
cv2.imshow("img2", img2)
cv2.imshow("bit_not1", bit_not1)
cv2.imshow("bit_not2", bit_not2)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
