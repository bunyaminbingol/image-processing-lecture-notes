```python
import cv2
import numpy as np

img = cv2.imread("veriler/helikopter.jpg", 0) # Sonuna 0 koyarak resmi siyah beyaz hale getiriyorum.
row, col = img.shape # Shape bana satr ve stun deerlerini veriyordu ben bu deerleri iki deiken ierisinde tutacam
print(row)
print(col)
```

```python
M = cv2.getRotationMatrix2D((col/2,row/2),90,1) # Bu fonksiyon iki boyutta bir rotasyon yani yn deitirme iine yaryor.
# ilk parametre stun ve satr saylarnn yarsn alyoruz
# 2. parametre ka derece dndreceimizi giriyoruz.
# 3. paramatre de ise lei giriyoruz, 1 girdim yani bire bir resim ekrana gelsin.

dst = cv2.warpAffine(img,M,(col,row)) # ve bu ilemi uyguluyoruz.

cv2.imshow("dst 90 derece", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
