```python
import cv2
import numpy as np

img = cv2.imread("veriler/helikopter.jpg", 0) # Sonuna 0 koyarak resmi siyah beyaz hale getiriyorum.
row, col = img.shape # Shape bana satır ve sütun değerlerini veriyordu ben bu değerleri iki değişken içerisinde tutacağım
print(row)
print(col)
```

```python
M = cv2.getRotationMatrix2D((col/2,row/2),90,1) # Bu fonksiyon iki boyutta bir rotasyon yani yön değiştirme işine yarıyor.
# ilk parametre sütun ve satır sayılarının yarısını alıyoruz
# 2. parametre kaç derece döndüreceğimizi giriyoruz.
# 3. paramatre de ise ölçeği giriyoruz, 1 girdim yani bire bir resim ekrana gelsin.

dst = cv2.warpAffine(img,M,(col,row)) # ve bu işlemi uyguluyoruz.

cv2.imshow("dst 90 derece", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
