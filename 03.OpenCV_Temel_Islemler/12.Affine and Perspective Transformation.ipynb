# Affine Tranformation
orijinal grntdeki tm paralel izgiler kt grntsnde yine paralel olacaktr. Dnm matrisini bulmak iin, girdi grntsnden  noktaya ve bunlarn kt grntsndeki karlk gelen konumlarna ihtiyacmz var. Daha sonra cv2.getAffineTransform geirilecek olan bir 2x3 matris oluturur cv2.warpAffine .


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("veriler/helikopter.jpg")
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# Perspective Transformation
Perspektif dnm iin 3x3 dnm matrisine ihtiyacnz var. Dz izgiler, dnmden sonra bile dz kalacaktr. Bu dnm matrisini bulmak iin, girdi grntsnde 4 noktaya ve kt grntsnde karlk gelen noktalara ihtiyacnz vardr. Bu 4 nokta arasnda 3 tanesi edorusal olmamaldr. Daha sonra dnm matrisi cv2.getPerspectiveTransform ilevi ile bulunabilir . Sonra bu 3x3 dnm matrisiyle cv2.warpPerspective uygulayn .


```python
img = cv2.imread("veriler/helikopter.jpg")
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
