# Affine Tranformation
orijinal görüntüdeki tüm paralel çizgiler çıktı görüntüsünde yine paralel olacaktır. Dönüşüm matrisini bulmak için, girdi görüntüsünden üç noktaya ve bunların çıktı görüntüsündeki karşılık gelen konumlarına ihtiyacımız var. Daha sonra cv2.getAffineTransform geçirilecek olan bir 2x3 matris oluşturur cv2.warpAffine .


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("veriler/helikopter.jpg")
rows,cols,ch = img.shape

pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[10,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

# Perspective Transformation
Perspektif dönüşümü için 3x3 dönüşüm matrisine ihtiyacınız var. Düz çizgiler, dönüşümden sonra bile düz kalacaktır. Bu dönüşüm matrisini bulmak için, girdi görüntüsünde 4 noktaya ve çıktı görüntüsünde karşılık gelen noktalara ihtiyacınız vardır. Bu 4 nokta arasında 3 tanesi eşdoğrusal olmamalıdır. Daha sonra dönüşüm matrisi cv2.getPerspectiveTransform işlevi ile bulunabilir . Sonra bu 3x3 dönüşüm matrisiyle cv2.warpPerspective uygulayın .


```python
img = cv2.imread("veriler/helikopter.jpg")
rows,cols,ch = img.shape

pts1 = np.float32([[56,65],[368,52],[28,387],[389,390]])
pts2 = np.float32([[0,0],[300,0],[0,300],[300,300]])

M = cv2.getPerspectiveTransform(pts1,pts2)

dst = cv2.warpPerspective(img,M,(300,300))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```
